Normal Distribution:                   
The normal distribution also known as the Gaussian Distribution, is a continuous probability distribution that is symmetric around its mean, whixh is also its median and mode. It is characterised by its bell-shaped curve. The distribution is completely deternined by two parameters: the mean () and the standard deviation ().

### **Probability Density Function (pdf):**
The probability density function is a function that describes the likelihood of a continuous random variable falling within a particular range.
For normal distribution                 
Here:                   
`x` is the variable.                           
`u` is the mean.                         
`sigma` is the standard deviation.                  
The integral of the pdf over the entire range of possible values is equal to 1, indicating that the probability of all possible outcomes is 1.

#### **Box and Whisker Plot**
It is useful for visualizing the spread and skew of data.
The red line represents the median of the data, and the box represents the bounds of the 1st and 3rd quartiles.
So, the half of the data exists within the box.
The dotted-line 'whiskers' indicate the range of the data except for outliers which are plotted outside the whiskers. 
Outliers are 1.5x or more the interquartile range.
This example below creates uniformly distributed random numbers between -40 and 60+ a few outliers above 100 and below -100:

```
       Q1-1.5IQR     Q1   median   Q3     Q3+1.5IQR
                     |------:-------|              
 o         |---------|      :       |--------|   o  o
                     |------:-------|
flier                <--------------->         fliers
                           IQR```

````python
import pandas as pd
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
df = pd.DataFrame({'data': data}, columns=['data'])
print(df)
df.describe()
````

````python
plt.boxplot(df['data'], labels=df.columns, vert=False, showfliers=True, meanline=True)
plt.show()
````

## **Introduction to Seaborn**
It is a Python data visualization library based on Matplotlib that provides a high-level interface for creating informative and attractive statistical graphics.

````python
import seaborn as sns
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')
````

````python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

sns.lineplot(x=x, y=y, linestyle=':', label=True, palette='Blues')

plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Chart')
plt.show()
````

````python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

sns.scatterplot(x=x, y=y, color='green', marker='o', edgecolor='black', legend='auto')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.show()
````

````python
category = ['a', 'b', 'c', 'd']
value = [3,8,5,12]
bar = sns.barplot(x=category, y=value, palette='Set3', orient='v', legend='auto')
bar.bar_label(bar.containers[1], fontsize=10, fontweight='bold', label_type='edge')
bar.plot('a', 4, '*', color='black', markersize=10)
plt.grid(axis='both')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot')
plt.show()
````

````python
# Sample data
categories = ['Category A', 'Category B', 'Category C']
values_1 = [3,8,5]
values_2 = [5,2,7]

# Creating a DataFrame for Seaborn
data = {'Categories': categories, 'Group 1': values_1, 'Group 2': values_2}
df = pd.DataFrame(data)
df
````

````python
sns.lineplot(x='Group 1', y = df['Group 2'], data = df, color = 'red', linestyle=':')
plt.grid(True)
plt.xlabel('Value 1')
plt.ylabel('Value 2')
plt.title('Line Plot')
plt.show()
````

````python
# Plotting bar graph using seaborn
sns.barplot(x='Categories', y='Group 1', data = df, color='lightgreen', label='Group 1')
sns.barplot(x='Categories', y='Group 2', data = df, color='darkgreen', bottom=values_1, label='Group 2')


plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Seaborn stacked bar graph with legend')
plt.xticks(rotation=90)

plt.show()
````
