```python
df.groupby('island')['sex'].sum()
```

```python
ax = sns.barplot(data=df, x='island', y ='sex', estimator='mean', errorbar=None)
ax.bar_label(ax.containers[0], fontsize=10)
ax.plot('Sun', 3.5, 'r*', markersize=10)
plt.grid()
plt.xlabel('Island')
plt.ylabel('Sex')
plt.title('Barplot of island')
plt.show()
```

```python
sns.countplot(data=df,x='island',hue='sex', palette='Set2')
plt.grid(True)
plt.title('Count of sex')
plt.show()
```

```python
sex = df['sex'].value_counts()
sex
```

```python
plt.pie(sex, labels=sex.index, autopct='%1.f%%', explode=[0, 0.1], colors=sns.color_palette('Set2'))
plt.legend()
plt.title('Distribution of Sex')
plt.show()
```

```python
sns.scatterplot(data=df, x='bill_length_mm', y='body_mass_g', alpha = 0.5, marker='*', edgecolor='black',legend=True,sizes=(20, 200)) #p,s,h,d,x,^,+,*,-
plt.title('Scatter Plot of Total Bill vs. body mass')
plt.show()
```

```python
sns.lineplot(data=df,x='species', y='body_mass_g',hue='sex', orient="x", marker='*', markersize=10, linestyle=':')
plt.title('Line Plot of sex')
plt.show()
```

```python
sns.boxplot(data=df, y='body_mass_g', fill=True, palette='Set2')
plt.title('Box Plot of Total body mass')
plt.show()
```

```python
sns.distplot(df['bill_length_mm'], hist=True, bins=30, color='b', kde=True, vertical=False)
plt.title('dtribution Plot of Total Bill length')
plt.show()
```

```python
sns.kdeplot(data=df, x='flipper_length_mm', hue="sex", shade=True)
plt.title('Kurnal density estimation Plot of flipper length')
plt.show()
```

```python
sns.histplot(data=df,x='bill_depth_mm',edgecolor= 'black', bins=10,element="step", color='blue') #poly , fill= False
plt.title('Hist Plot of Total Bill depth')
plt.show()
```

```python
sns.pairplot(df, hue='species', diag_kind="hist")
plt.suptitle('Pairplot of species', y=1.02, x=0.5)
plt.show()
```

```python
sns.violinplot(data=df, y='flipper_length_mm', split=False, inner_kws=dict(box_width=15, whis_width=2), palette='Set2')
plt.title('Violin Plot of Total flipper length', color='blue', fontsize=16, fontweight='bold')
plt.show()
```

```python
x= 'bill_length_mm'
y = 'body_mass_g'
sns.jointplot(x=x, y=y, kind='scatter', data = df, color='red') #“scatter” | “kde” | “hist” | “hex” | “reg” | “resid”
plt.show()
```

```python
sns.set_style("whitegrid")

# Create a swarmplot
sns.swarmplot(x='species', y='bill_length_mm',hue='body_mass_g', data=df, size=10, legend=True, marker='*')  # size is used for marker size , marker='*
plt.legend(loc='upper left')
# Show the plot
plt.show()
```
