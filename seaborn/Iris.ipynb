```python
ax = sns.barplot(data=df, x='species', y ='sepal_width', estimator='mean', errorbar=None)
ax.bar_label(ax.containers[0], fontsize=10)
ax.plot('versicolor', 3.0, 'r*', markersize=10)
plt.grid()
plt.xlabel('species')
plt.ylabel('sepal_width')
plt.title('Barplot of species')
plt.show()
```

```python
sns.countplot(data=df,x='petal_length',hue='species', palette='Set2')
plt.grid(True)
plt.title('Count of species')
plt.show()
```

```python
Species = df['species'].value_counts()
Species
```

```python
plt.pie(Species,labels=Species.index, autopct='%1.f%%', explode=[0,0.1,0], colors=sns.color_palette('Set2'))
plt.legend()
plt.title('Distribution of Species')
plt.show()
```

```python
sns.scatterplot(data=df, x='species', y='sepal_length', alpha = 0.5, marker='*', edgecolor='black',legend=True, sizes=(20, 200))
plt.title('Scatter Plot of species')
plt.show()
```

```python
sns.lineplot(data=df,x='sepal_length', y='sepal_width',hue='species', orient="x", marker='*', markersize=10, linestyle=':')
plt.title('Line Plot of Sepal length')
plt.show()
```

```python
sns.boxplot(data=df, y='petal_length', fill=True, palette='Set2')
plt.title('Box Plot of petal length')
plt.show()
```

```python
sns.distplot(df['sepal_width'], hist=True, bins=30, color='b', kde=True, vertical=False)
plt.title('dtribution Plot of sepal width')
plt.show()
```

```python
sns.kdeplot(data=df, x='petal_length', hue="species", shade=True)
plt.title('Kurnal density estimation of species')
plt.show()
```

```python
sns.histplot(data=df,x='sepal_width',edgecolor= 'black', bins=10,element="step", color='blue') #poly , fill= False
plt.title('Hist Plot of sepal')
plt.show()
```

```python
#sns.pairplot(df , kind = "scatter")
sns.pairplot(df, hue='species', diag_kind="hist")
plt.suptitle('Pairplot of Numerical Variables by species', y=1.02, x=0.5)
plt.show()
```

```python
Num = df.select_dtypes(include=['int','float']).columns.tolist()
Num
```

```python
correlation_matrix = df[Num].corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='PuBu')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.violinplot(data=df, y='sepal_width', split=False, inner_kws=dict(box_width=15, whis_width=2), palette='Set2')
plt.title('Violin Plot of sepal', color='blue', fontsize=16, fontweight='bold')
plt.show()
```

```python
x= 'petal_length'
y = 'sepal_width'
sns.jointplot(x=x, y=y, kind='scatter', data = df, color='red') 
plt.show()
```

```python
sns.set_style("whitegrid")


sns.swarmplot(x='petal_length', y='sepal_width',hue='species', data=df, size=10, legend=True, marker='*')  
plt.legend(loc='upper left')

plt.show()
```
