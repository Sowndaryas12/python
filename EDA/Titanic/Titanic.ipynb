we will manipulate the variables related to age in our dataset.upon examining the sample data 

```python
def categorize(age):
    if age< 17.0:
        return 'Young'
    elif age<34.0:
        return 'Adult'
    elif age< 54.0:
        return 'Middle age'
    else:
        return 'Seniors'
```

```python
df['Age_group']= df['age'].apply(categorize)
```
