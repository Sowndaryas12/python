```python
#22. Visualize the distribution of BasePay, OvertimePay, and OtherPay using histograms or kernel density plots
plt.figure(figsize=(15, 5))


plt.subplot(1, 3, 1)
sns.histplot(df2['BasePay'].dropna(), bins=30, kde=False, color='blue')
plt.title('BasePay Distribution')
plt.xlabel('BasePay')


plt.subplot(1, 3, 2)
sns.histplot(df2['OvertimePay'].dropna(), bins=30, kde=False, color='green')
plt.title('OvertimePay Distribution')
plt.xlabel('OvertimePay')


plt.subplot(1, 3, 3)
sns.histplot(df2['OtherPay'].dropna(), bins=30, kde=False, color='red')
plt.title('OtherPay Distribution')
plt.xlabel('OtherPay')


plt.tight_layout()
plt.show()
```
